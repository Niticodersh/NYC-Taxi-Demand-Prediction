```python
columns = df.columns.tolist()
```

```python
columns.remove('time')
columns.remove('mean_passenger_per_trip')
```

```python
columns
```

```python
max_values = [600*1e3,50,35,100*1e3,100*1e3,10,53*1e3]
```

```python
for i in range(len(columns)):
    df[columns[i]] = (df[columns[i]]-df[columns[i]].min())/(max_values[i]-df[columns[i]].min())
```

```python
df.head()
```

```python
for c in columns:
    df[c] = np.where(df[c]>1,1,df[c])
```

```python
df.to_parquet('normalised_data.parquet')
```
